### First regression model used: SIMPLE LINEAR REGRESSION

```python
# selecting the relevant columns
data = data[['weight', 'price']]  #selecting relevant features to be used for the simple linear regression. 'weight' is the independent variable, while 'price' is the dependent variable
```

```python
#handling missing values
data.dropna()  
```

```python
#the S.I units in the column 'weight' has to be removed, to properly make weight look like a numeric type. Also, some columns have multiple enteries, so we have to split these entries. and finally, convert the weights to a single unit, for consistency and accuracy

# a function to convert weight strings to float in kg
def convert_weight(weight_str):
    try:
        weight_str = weight_str.lower().strip()
        
        total_weight_kg = 0.0
        
        parts = weight_str.split()
        
        for i in range(len(parts)):
            part = parts[i]
            if part.isdigit() or ('.' in part and part.replace('.', '', 1).isdigit()):
                weight_value = float(part)
                unit = parts[i + 1] if i + 1 < len(parts) else ''
                
            if 'pound' in unit:
                total_weight_kg += weight_value * 0.4535
            elif 'oz' in unit or 'ounce' in unit:
                total_weight_kg += weight_value * 0.0283
            elif 'kg' in unit:
                total_weight_kg += weight_value
            elif 'lb' in unit:
                total_weight_kg += weight_value * 0.4535
            
        return total_weight_kg if total_weight_kg > 0 else None
    except Exception as e:
        print(f"conversion failed for value: {weight_str}, error:{e}")
        return None
data.loc[:, 'weight'] = data['weight'].apply(convert_weight)    
```

```python
#init. GridSearchCv for both models
ridge_grid = GridSearchCV(ridge, ridge_params, cv = 5, scoring = 'neg_mean_squared_error')

lasso_grid = GridSearchCV(lasso, lasso_params, cv = 5, scoring = 'neg_mean_squared_error')
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

# Define the parameter grid for Gradient Boosting
gb_params = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 5, 10]
}

# Initialize Gradient Boosting regressor
gb = GradientBoostingRegressor(random_state=42)

# Initialize GridSearchCV
gb_grid = GridSearchCV(gb, gb_params, cv=5, scoring='neg_mean_squared_error')

# Fit GridSearchCV
gb_grid.fit(X_train, y_train)

# Get the best estimator
best_gb = gb_grid.best_estimator_

# Predict using the best estimator
gb_pred = best_gb.predict(X_test)

# Evaluate the model
gb_mse = mean_squared_error(y_test, gb_pred)
gb_r2 = r2_score(y_test, gb_pred)
print(f'Gradient Boosting - Best Parameters: {gb_grid.best_params_}')
print(f'Gradient Boosting - Mean Squared Error: {gb_mse}')
print(f'Gradient Boosting - R-squared: {gb_r2}')

```
